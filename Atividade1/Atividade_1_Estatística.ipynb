# <font color='orange'> **Inclusão de bibliotecas necessárias**

```python
import matplotlib.pyplot as plt       # plotagem de gráficos
import pandas as pd                   # manipulação e análise do dataset
import numpy as np                    # manipulação de vetores
import seaborn as sns                 # plotagem e análises estatísticas
import os                             # interação com o sistema computacional
import kagglehub                      # importação do dataset(csv)
from IPython.display import display   # uso do diplay para imprimir tabelas
```

# <font color='orange'> **Download do dataset**

<a href="https://www.kaggle.com/datasets/alanjo/ddr2ddr3ddr4ddr5-ram-benchmarks" target="_blank">Origem do dataset</a>

```python
# Download da úlltima versão do dataset
path = kagglehub.dataset_download("alanjo/ddr2ddr3ddr4ddr5-ram-benchmarks")

# Obtenção dos arquivos do diretório do dataset
arqs = os.listdir(path)

# Print do diretório do dataset e arquivos pertencentes
print("Path to dataset files:", path)
for arqv in arqs:
  print(arqv)
```

# <font color='orange'> **Criação do dataframe principal**

```python
# Criação do dataframe principal a partir do arquivo csv com os dados do dataset
df = pd.read_csv(path + '/RAM_Benchmarks_megalist.csv')

# Imprime dimensões da tabela de dados
print('Dimensão:', df.shape)

# Imprime 30 primeiras linhas da tabela tabela de dados
display(df.head(30))
```

# <font color='orange'> **Análise do dataset**

## <font color='orange'>1) Análise da relação entre a latência de operação (ns) e a velocidade de escrita (MB/s)

### a) Correlação

```python
# Obtenção dos coeficientes de correlação (Pearson, Spearman e Kendall) entre latência e velocidade de escrita
pearson_corr_latency_write = df['latency'].corr(df['write'], method='pearson')
spearman_corr_latency_write = df['latency'].corr(df['write'], method='spearman')
kendall_corr_latency_write = df['latency'].corr(df['write'], method='kendall')

# Vetores de linhas da tabela
array_names = ['rp (coeficiente de Pearson)', 'rs (coeficiente de Spearman)', 'rk (coeficiente de Kendall)']
array_corr_1 = np.array([pearson_corr_latency_write,
                        spearman_corr_latency_write, kendall_corr_latency_write])
array_corr_1_reshape = array_corr_1.reshape(1, 3)  # inversão de dimensão para corrigir erro de formação da tabela pelo pandas

# Obtenção do data frame para impressão da tabela de coeficientes
df1 = pd.DataFrame(array_corr_1_reshape, columns=array_names)

# Impressão da tabela de coeficientes no formato html
display(HTML(df1.to_html(index=False)))

```

<center>
<font size="4"><p>
  Tabela de referência para interpretação do coeficiente de correlação entre variáveis quantitativas
</p>

</center>

<font size="4"><p>
  &nbsp;&nbsp;&nbsp;&nbsp; Analisando os coeficientes de Pearson ($r_{p}$), Spearman ($r_{s}$) e Kendall ($r_{k}$), e correspondendo seus valores com os itervalos da tabela de interpretação desses coeficientes, é possível notar velocidade de escrita na memória RAM e a latência estão negativamente relacionados e têm uma correlação intermediária.
</p>


### b) Covariância

```python
# Cálculo e impressão da covariância entre latência de operação e a velocidade escrita na RAM
cov_latency_write = df['latency'].cov(df['write'])
print('Covariância:', cov_latency_write)
```

&nbsp;&nbsp;&nbsp;&nbsp; A partir do sinal negativo obtido do valor covariância, podemos observar que, de forma concordante com a correlação, as variáveis de latência e velocidade de escrita estão relacionadas negativamente, ou seja, uma tende a diminuir com o aumento da outra

&nbsp;&nbsp;&nbsp;&nbsp; Por outro lado, analisar o módulo não tem grande utilidade para determinar a relação entre as variáveis, já que o mesmo é influenciado pelas unidades de medida das variáveis quantitativas envolvidas em seu cálculo.

### c) Gráfico de dispersão

```python
# Plotagem gráfico do gráfico de dispersão com linha de tendência
sns.lmplot(
    x='write',
    y='latency',
    data=df,
    scatter_kws={'color': 'blue', 's':20},
    line_kws={'color': 'red'}
)
plt.xlabel('Escrita [MB/s]')
plt.ylabel('Latência [ns]')
plt.title('Gráfico de dispersão: Latência de operação da RAM vs Velocidade de escrita na RAM')
plt.show()
```

## <font color='orange'>2) Análise da relação entre latência de operação (ns) e a velocidade de leitura (MB/s)

### a) Correlação

```python
# Obtenção dos coeficientes de correlação (Pearson, Spearman e Kendall) entre latência e velocidade de leitura
pearson_corr_latency_readUncached = df['latency'].corr(df['readUncached'], method='pearson')
spearman_corr_latency_readUncached = df['latency'].corr(df['readUncached'], method='spearman')
kendall_corr_latency_readUncached = df['latency'].corr(df['readUncached'], method='kendall')

# Vetores de linhas da tabela
array_names = ['rp (coeficiente de Pearson)', 'rs (coeficiente de Spearman)', 'rk (coeficiente de Kendall)']
array_corr_2 = np.array([pearson_corr_latency_readUncached,
                         spearman_corr_latency_readUncached, kendall_corr_latency_readUncached])
array_corr_2_reshape = array_corr_2.reshape(1, 3) # inversão da dimensão para corrigir erro de formação da tabela pelo pandas

# Obtenção do data frame para impressão da tabela de coeficientes
df2 = pd.DataFrame(array_corr_2_reshape, columns=array_names)

# Impressão da tabela de coeficientes no formato html
display(HTML(df2.to_html(index=False)))
```

<font size="4"><p>
 &nbsp;&nbsp;&nbsp;&nbsp; Por outro lado, considerando os coeficientes de Spearman ($r_{s}$) e Kendall ($r_{k}$), e relacionando seus valores com os itervalos da tabela de interpretação desses coeficientes, é possível notar que a velocidade de leitura da memória RAM e a latência estão negativamente relacionados e têm uma correlação intermediária.

### b) Covariância

```python
# Cálculo e impressão da covariância entre latência de operação e a velocidade de leitura direta da RAM
cov_latency_readUncached = df['latency'].cov(df['readUncached'])
print('Covariância:', cov_latency_readUncached)
```

&nbsp;&nbsp;&nbsp;&nbsp; A partir do sinal negativo obtido do valor covariância, podemos observar que, de forma concordante com a correlação, as variáveis de latência e velocidade de leitura estão relacionadas negativamente, ou seja, uma tende a diminuir com o aumento da outra

### c) Gráfico de dispersão

```python
# Plotagem gráfico do gráfico de dispersão com linha de tendência
sns.lmplot(
    x='readUncached',
    y='latency',
    data=df,
    scatter_kws={'color': 'blue', 's':20},
    line_kws={'color': 'red'}
)
plt.xlabel('Leitura direta [MB/s]')
plt.ylabel('Latência [ns]')
plt.title('Gráfico de dispersão: Latência de operação da RAM vs Velocidade de leitura da RAM')
plt.show()
```

## <font color='orange'>3) Análise da relação entre a velocidade escrita (MB/s) e a velocidade de leitura (MB/s)

&nbsp;&nbsp;&nbsp;&nbsp; Por outro lado, analisar o módulo não tem grande utilidade para determinar a relação entre as velocidades, já que o mesmo é influenciado pelas unidades de medida das variáveis quantitativas envolvidas em seu cálculo.

## <font color='orange'> 4) Análise final do dataset

### Heatmap de correlações pelo método de Pearson

```python
# Plotagem do heatmap de correlações entre variáveis quantitativas do dataset através do método de Pearson
sns.heatmap(
    df[['latency', 'write', 'readUncached']].corr(method='pearson'),
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=.5,
    cbar=True,
    square=True
)
plt.title('Heatmap de correlação entre variávei quantitativas')
plt.show()
```

  </p>
  
  <font size="4"><p>
  &nbsp;&nbsp;&nbsp;&nbsp; Com isso, apesar de a velocidade de leitura ser a mais determinante na eficiência de uma memória RAM, a velocidade de escrita é ainda extremamente relevante para essa eficiência e, consequentemente, ambas as taxas devem ser consideradas no momento de comprar componentes de um sistema computacional de qualidade.
